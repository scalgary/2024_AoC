```python
def read_and_split_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    # Split the content into two parts based on an empty line
    parts = content.split('\n\n')
    if len(parts) != 2:
        raise ValueError("The file does not contain exactly one empty line to split on.")

    part1 = parts[0].strip()
    part2 = parts[1].strip()

    return part1, part2

def parse_relations(relations_str):
    relations = {}
    for line in relations_str.split('\n'):
        a, b = map(int, line.split('|'))
        if a not in relations:
            relations[a] = []
        relations[a].append(b)
    return relations

def parse_lists(lists_str):
    lists = []
    for line in lists_str.split('\n'):
        elements = list(map(int, line.split(',')))
        lists.append(elements)
    return lists

# Lire et diviser le fichier
file_path = 'test05.txt'
file_path = 'input05.txt'
relations_str, lists_str = read_and_split_file(file_path)

# Parser les relations
relations = parse_relations(relations_str)

print(relations)

# Parser les listes
lists = parse_lists(lists_str)

print("Relations:", relations)
print("Listes:", lists)


```

```python
def check_is_relation(number1,number2):
    if number1 in relations.keys():
        return number2 in relations[number1]
    else:
        return False
```

```python
# Ouvrir le fichier en mode lecture
file_path = "test06.txt"
file_path = "input06.txt"
with open(file_path, 'r') as file:
    # Lire toutes les lignes du fichier
    lines = file.readlines()

# Convertir chaque ligne en une liste de caractères et stocker dans une matrice
matrix = [list(line.strip()) for line in lines]

# Afficher la matrice
for row in matrix:
    print(row)
```

```python
#where is the ^

def find_element_coordinates(matrix, element):
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value == element:
                return (i, j) #if unique element
    return None

def add_tuples(tuple1, tuple2):
 
    return (tuple1[0] + tuple2[0], tuple1[1] + tuple2[1])



    
find_element_coordinates(matrix,"^")

def move_to_next_key(current_key, key_order):
    
    current_index = key_order.index(current_key)
    next_index = (current_index + 1) % len(key_order)
    return key_order[next_index]

key_order = ["up", "right", "down", "left"]
move_to_next_key("up",key_order)
```

```python
def get_element(matrix, coordinates):
    row, col = coordinates
    return matrix[row][col]
```

```python
def move(start,direction):
    key_order = ["up", "right", "down", "left"]
    possible_direction = {"up" :(-1,0),"right" :(0,1), "down" :(1,0), "left" : (0,-1)} #up right down left
    possible = add_tuples(start,possible_direction[direction])
    # possible is in matrix?
    if get_element(matrix, possible)=="#":
        return start,move_to_next_key(direction, key_order)
    else:
        return possible,direction

    

    
```

```python
def is_on_edge(coordinates):
    x,y = coordinates
    if x==0 or x == len(matrix):
        return True
    if y==0 or y == len(matrix[0]):
        return True
    return False
       
```

```python
def is_on_edge(position):
    rows = len(matrix)
    cols = len(matrix[0]) if rows > 0 else 0
    row, col = position
    return row == 0 or row == rows - 1 or col == 0 or col == cols - 1
```

```python
# Trouver les coordonnées de l'élément "^"
start = find_element_coordinates(matrix, "^")

if start:
    # Remplacer l'élément à la position (i, j) par "X"
    matrix[start[0]][start[1]] = "X"

direction = "up"
```

```python
def count_x(matrix):
    count = 0
    for row in matrix:
        for element in row:
            if element == "X":
                count += 1
    return count
```

```python
# Ouvrir le fichier en mode lecture
file_path = "test08.txt"
file_path = "input08.txt"
with open(file_path, 'r') as file:
    # Lire toutes les lignes du fichier
    lines = file.readlines()

# Convertir chaque ligne en une liste de caractères et stocker dans une matrice
matrix = [list(line.strip()) for line in lines]

# Afficher la matrice
for row in matrix:
    print(row)
```

```python
#get unique value.

from os import uname


unique_values = set()
for row in matrix:
    for element in row:
        unique_values.add(element)
unique_values.remove(".")

print(unique_values)  # Output: {1, 2, 3, 4, 6, 7, 8}
```

```python


# Créer un dictionnaire vide à partir des valeurs uniques
value_to_coordinates = {value: [] for value in unique_values}

# Remplir le dictionnaire avec les coordonnées des valeurs 'A' et '0'
for i, row in enumerate(matrix):
    for j, value in enumerate(row):
        if value in value_to_coordinates.keys():
            value_to_coordinates[value].append((i, j))

print(value_to_coordinates)
# Output: {'.': [], 'A': [(0, 0), (1, 1), (2, 2)], '0': [(0, 3), (2, 0), (3, 3)]}
```

```python
def check_criteria(mylist):
    mylist_int =[int(x) for x in mylist]
    result = [ x -y for x,y in zip(mylist_int[1:],mylist_int[0:])]
    allincreasingordecreasing = all(x>0 for x in result) | all(x<0 for x in result)
    nottoomuch = all([1<= abs(x) <4 for x in result])
    return allincreasingordecreasing and nottoomuch
```

```python
def modify_list(mylist):
    new_list= mylist.copy()
    mylist_int =[int(x) for x in new_list]
    result = [ x -y for x,y in zip(mylist_int[1:],mylist_int[0:])]
    # result all same sign
    positif = [ x for x in result if x>0]
    negatif = [ x for x in result if x<0]
    if len(positif) ==1:
        del new_list[result.index(positif[0])+1]
        return new_list
    if len(negatif) ==1:
        del new_list[result.index(negatif[0])+1]
        return new_list
    if len(positif) + len(negatif) < len(result):
        del new_list[result.index(0)+1]
        return new_list
    
    return new_list
```

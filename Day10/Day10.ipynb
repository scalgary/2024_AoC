```python
class Node:
    def __init__(self, value, x, y):
        self.value = value
        self.x = x
        self.y = y
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

    def __str__(self):
        return f"Node(value={self.value}, x={self.x}, y={self.y})"

def get_neighbors(matrix, x, y):
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right
    neighbors = []
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < len(matrix) and 0 <= ny < len(matrix[0]):
            neighbors.append((nx, ny))
    return neighbors

def add_children(node, matrix, generation):
    neighbors = get_neighbors(matrix, node.x, node.y)
    for nx, ny in neighbors:
        if matrix[nx][ny] == str(generation):
            child_node = Node(matrix[nx][ny], nx, ny)
            node.add_child(child_node)
            add_children(child_node, matrix, generation + 1)

def build_tree(matrix):
    root = Node('root', -1, -1)
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value == '0':
                node = Node(value, i, j)
                root.add_child(node)
                add_children(node, matrix, 1)
    return root

def collect_branches(node, path=[], branches=[]):
    path = path + [node]
    if not node.children:
        branches.append(path)
    for child in node.children:
        collect_branches(child, path, branches)
    return branches

def count_branches_with_0_and_9(branches):
    count = 0
    for branch in branches:
        has_0 = any(node.value == '0' for node in branch)
        has_9 = any(node.value == '9' for node in branch)
        if has_0 and has_9:
            count += 1
    return count

# Lire la matrice Ã  partir du fichier
file_path = "input10.txt"
with open(file_path, 'r') as file:
    lines = file.readlines()

# Convertir chaque ligne en une liste de caractÃ¨res et stocker dans une matrice
matrix = [list(line.strip()) for line in lines]

# Construire l'arbre Ã  partir de la matrice
root = build_tree(matrix)

# Collecter les branches
branches = collect_branches(root)

# Imprimer les branches
for branch in branches:
    print(" -> ".join(str(node) for node in branch))

# Compter les branches avec un nÅud de valeur 0 et un nÅud de valeur 9
branches_with_0_and_9_count = count_branches_with_0_and_9(branches)
print("Nombre de branches avec un nÅud de valeur 0 et un nÅud de valeur 9:", branches_with_0_and_9_count)
```

```python
def check_word(matrix, word):
    rows = len(matrix)
    cols = len(matrix[0]) if rows > 0 else 0
    word_len = len(word)

    # Check horizontally
    for i in range(rows):
        for j in range(cols - word_len + 1):
            if ''.join(matrix[i][j:j + word_len]) == word:
                return True

    # Check vertically
    for i in range(rows - word_len + 1):
        for j in range(cols):
            if ''.join(matrix[i + k][j] for k in range(word_len)) == word:
                return True

    # Check diagonally (top-left to bottom-right)
    for i in range(rows - word_len + 1):
        for j in range(cols - word_len + 1):
            if ''.join(matrix[i + k][j + k] for k in range(word_len)) == word:
                return True

    # Check diagonally (top-right to bottom-left)
    for i in range(rows - word_len + 1):
        for j in range(word_len - 1, cols):
            if ''.join(matrix[i + k][j - k] for k in range(word_len)) == word:
                return True

    return False

# Exemple de matrice


# Vrifier si le mot "XMAS" apparat dans la matrice
word = "XMAS"
found = check_word(matrix, word)
print(f"Le mot '{word}' apparat dans la matrice:", found)
```

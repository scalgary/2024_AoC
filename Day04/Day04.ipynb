```python
# Ouvrir le fichier en mode lecture
file_path = "test04.txt"
file_path = "input04.txt"
with open(file_path, 'r') as file:
    # Lire toutes les lignes du fichier
    lines = file.readlines()

# Convertir chaque ligne en une liste de caractères et stocker dans une matrice
matrix = [list(line.strip()) for line in lines]

# Afficher la matrice
for row in matrix:
    print(row)
```

```python
class Node:
    def __init__(self, value, x, y):
        self.value = value
        self.x = x
        self.y = y
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

    def __repr__(self):
        return f"Node({self.value}, ({self.x}, {self.y}))"

def get_neighbors(matrix, x, y):
    neighbors = []
    rows = len(matrix)
    cols = len(matrix[0]) if rows > 0 else 0

    # Liste des déplacements possibles (haut, bas, gauche, droite)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < rows and 0 <= ny < cols:
            neighbors.append((nx, ny))

    return neighbors

def add_children(node, matrix, generation):
    if generation == 1:
        target_value = 'M'
        next_generation = 2
    elif generation == 2:
        target_value = 'A'
        next_generation = 3
    elif generation == 3:
        target_value = 'S'
        next_generation = 4
    else:
        return

    neighbors = get_neighbors(matrix, node.x, node.y)
    for nx, ny in neighbors:
        if matrix[nx][ny] == target_value:
            child_node = Node(matrix[nx][ny], nx, ny)
            node.add_child(child_node)
            add_children(child_node, matrix, next_generation)

def build_tree(matrix):
    root = Node('root', -1, -1)
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value == 'X':
                node = Node(value, i, j)
                root.add_child(node)
                add_children(node, matrix, 1)
    return root

def print_tree(node, level=0):
    print(' ' * level * 2 + str(node))
    for child in node.children:
        print_tree(child, level + 1)

def get_branch_names(node, current_path=""):
    if not node.children:
        return [current_path + node.value]
    branch_names = []
    for child in node.children:
        branch_names.extend(get_branch_names(child, current_path + node.value))
    return branch_names

# Exemple de matrice


# Construire l'arbre à partir de la matrice
root = build_tree(matrix)

# Afficher la structure de l'arbre
print_tree(root)

# Obtenir les noms des branches de l'arbre sans le mot "root"
branch_names = []
for child in root.children:
    branch_names.extend(get_branch_names(child))

print("Noms des branches de l'arbre:", branch_names)
resultat = [col for col in branch_names if col == "XMAS"]
len(resultat)
```

```python
def count_word_occurrences(matrix, words):
    rows = len(matrix)
    cols = len(matrix[0]) if rows > 0 else 0
    count = 0

    for word in words:
        word_len = len(word)

        # Check horizontally
        for i in range(rows):
            for j in range(cols - word_len + 1):
                if ''.join(matrix[i][j:j + word_len]) == word:
                    count += 1

        # Check vertically
        for i in range(rows - word_len + 1):
            for j in range(cols):
                if ''.join(matrix[i + k][j] for k in range(word_len)) == word:
                    count += 1

        # Check diagonally (top-left to bottom-right)
        for i in range(rows - word_len + 1):
            for j in range(cols - word_len + 1):
                if ''.join(matrix[i + k][j + k] for k in range(word_len)) == word:
                    count += 1

        # Check diagonally (top-right to bottom-left)
        for i in range(rows - word_len + 1):
            for j in range(word_len - 1, cols):
                if ''.join(matrix[i + k][j - k] for k in range(word_len)) == word:
                    count += 1

    return count



# Vérifier le nombre de fois que les mots "XMAS" et "SAMX" apparaissent dans la matrice
words = ["XMAS", "SAMX"]
occurrences = count_word_occurrences(matrix, words)
print(f"Les mots {words} apparaissent {occurrences} fois dans la matrice.")
```
